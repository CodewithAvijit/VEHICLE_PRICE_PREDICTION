```python
data= pd.read_csv('dataset_final.csv')
x=data.iloc[:, :-1]
y=data['price']
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.20, random_state=42)
data.info()
```

```python
model=XGBRegressor(objective="reg:squarederror",booster='gbtree',max_depth=3,n_estimators=200)
model.fit(xtrain,ytrain)
```

```python
print("TRAIN SCORE: ",model.score(xtrain,ytrain),"TEST SCORE: ",model.score(xtest,ytest))
```

```python
bestmin = float('inf')
bestindex=0
for i in range(2,80):
    xg=XGBRegressor(max_depth=i)
    xg.fit(xtrain,ytrain)
    score=mean_absolute_error(ytest,xg.predict(xtest))
    if score<bestmin:
        bestmin=score
        bestindex=i
        
print("BEST SCORE: ",bestmin,bestindex)
```

```python
ypred=model.predict(xtest)
print("MAE: ",mean_absolute_error(ytest,ypred))
print("MSE ",mean_squared_error(ytest,ypred))
print("R2score: ",r2_score(ytest,ypred)*100)
print("RMSE: ",root_mean_squared_error(ytest,ypred)*100)
```
